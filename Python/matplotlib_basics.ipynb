````python
import matplotlib.pyplot as plt
import numpy as np
````

# Simple plot of two functions

```python
fig = plt.figure()
```
The top level container for all the plot elements. Can add axes (different face color, projections).

The Axes contains most of the figure elements: Axis, Tick, Line2D, Text, Polygon, etc., and sets the coordinate system. We can get both 



````python
#       predefined styles (classic, bmh, defauls)
# plt.style.use('default')

#       here we update some parameters for the plot. Note: matplotlibrc is actually a file somewhere.
#       rc = "run commands". It is used for any file that contains startup information for a command.
newParams = {'figure.figsize'  : (12, 6),  # Figure size
             'figure.dpi'      : 200,      # figure resolution
             'axes.titlesize'  : 20,       # fontsize of title
             'axes.labelsize'  : 25,       # fontsize of axes labels
             'axes.linewidth'  : 2,        # width of the figure box lines
             'lines.linewidth' : 2,        # width of the plotted lines
             'savefig.dpi'     : 200,      # resolution of a figured saved using plt.savefig(filename)
             'ytick.labelsize' : 13,       # fontsize of tick labels on y axis
             'xtick.labelsize' : 13,       # fontsize of tick labels on x axis
             'legend.fontsize' : 18,       # fontsize of labels in legend
             'legend.frameon'  : True,     # activate frame on lengend?
            }
plt.rcParams.update(newparams)

#       data to plot
theta = np.linspace(0, 2*np.pi, 300, endpoint= True)
y1     = np.sin(theta**2)
y2     = np.cos(theta**2)
````

````python
# plt.figure(figsize=(16, 8)) # can specify a whole lot in here
plt.plot(theta, y1, label = r'sin($\theta ^2$)') # 'r' means raw string, otherwise '\' screw things up.
plt.plot(theta, y2, label = r'cos($\theta ^2$)')

plt.ylabel('Values [-]', fontsize=17) 
plt.xlabel(r'$\theta$ [rad]', fontsize=17)   

plt.xlim([0,8])
plt.ylim([-1.1, 1.1])

plt.xticks([0, np.pi, 2*np.pi], [0, r'$\pi$', r'$2\pi$'], fontsize=15)
# locs, labels = xticks() # can also get locations and labels
#       check ticker API for more: plt.xaxis.set_major_locator(xmajor_locator)

plt.title('Title of plot')   
plt.legend(loc = 'lower right', fontsize = 17)  # 'best' is default in rc-parameters (legend.loc)
plt.grid()                   
plt.show()
````
